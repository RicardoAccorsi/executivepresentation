```python
fig,ax = plt.subplots(
            figsize = (16,4) # Tamanho da figura
        )

wid = 0.2

ax.bar(ano_mes[ano_mes.Ano == 2015].Mês.values - 1.5*wid,
       ano_mes[ano_mes.Ano == 2015].Sales.values,
       color="#84ba5b", # Cor das barras
       width=wid,
       label="2015"
      )
ax.bar(ano_mes[ano_mes.Ano == 2016].Mês.values - wid/2,
       ano_mes[ano_mes.Ano == 2016].Sales.values,
       color="#00a9b5", # Cor das barras
       width=wid,
       label="2016"
      )
ax.bar(ano_mes[ano_mes.Ano == 2017].Mês.values + wid/2,
       ano_mes[ano_mes.Ano == 2017].Sales.values,
       color="#008ac5", # Cor das barras
       width=wid,
       label="2017"
      )
ax.bar(ano_mes[ano_mes.Ano == 2018].Mês.values + 1.5*wid,
       ano_mes[ano_mes.Ano == 2018].Sales.values,
       color="#0069c0", # Cor das barras
       width=wid,
       label="2018"
      )

for i in range(0,12):
    ax.annotate('{:,.0f}k'.format(ano_mes[ano_mes.Ano == 2018].Sales.values[i]/1000), # Valores que vão ser mostrados como rótulos, já formatados
                (ano_mes[ano_mes.Ano == 2018].Mês.values[i] + 1.5*wid, ano_mes[ano_mes.Ano == 2018].Sales.values[i]), # Posição (x,y que esses valores vão estar)
                ha="center",  # Alinhamento horizontal do texto
                va="top", # Alinhamento vertical do texto
                xytext=(5,13), # Deslocamento em relação ao ponto que o texto estaria
                textcoords="offset points",  # Sistema de coordenadas que o xytext vai ter como referência (nesse caso será em pontos, não em pixels)
                fontsize=10, # Tamanho da fonte dos rótulos
                fontweight='bold', # Espessura da fonte dos rótulos
                color="#0069c0" # Cor dos rótulos
               )

ax.legend()    

ax.set_ylim(0,130000) # melhorando dimensões do gráfico

# ajeitando labels do eixo x
ax.xaxis.set_ticks(range(1, 13)) 
ax.set_xticklabels(['Jan','Fev','Mar','Abr','Mai','Jun','Jul','Ago','Set','Out','Nov','Dez'])
ax.tick_params(axis='x',labelsize=14) # Melhorando o visual do eixo x

ax.set_xlim(0.5,12.6) # melhorando dimensões do gráfico

ax.yaxis.set_visible(False) # Retirando os valores do eixo y

ax.spines['top'].set_visible(False) # Retirando os traços do gráfico
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
    
plt.show()
```
